```python
import pandas as pd  #to read csv file
from sklearn.model_selection import train_test_split #used to divide the data to train and test sets
from sklearn.linear_model import LinearRegression #using linear regression model
import matplotlib.pyplot as plt #making plots and graphs
```

```python
dataset = pd.read_csv("weight_data.csv") #read data from the csv file
x = dataset.iloc[:,:-1].values #create a dataframe, taking height
y = dataset.iloc[:,1].values #create a dataframe, taking weight
print(dataset)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=1/5) #20% for train and 80%for test
print("x_train: ",x_train) #print all
print("x_test: ",x_test)
print("y_train: ",y_train)
print("y_test: ",y_test)
```

```python
plt.scatter(x,y,color="green") #draw a plot
plt.title('simple Linear Regression')
plt.xlabel('Height in Feets')
plt.ylabel('Weight in KGs')
plt.show()
```

```python
model = LinearRegression() #create a model
model.fit(x_train,y_train) #train the model using fit
print(model.coef_) #print coefficient
```

```python
plt.scatter(x, y, color = 'red') #plot the graph
plt.plot(x, model.predict(x), color = 'green')
plt.title('Simple Linear Regression')
plt.xlabel('Height in Feets')
plt.ylabel('Weight in KGs')
plt.show()
```

```python
pred_result = model.predict(x_test) #predict the result
```

```python
print("Test Height: ",x_test)
print("prediction Weight :",pred_result)
```
